```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python

df=pd.read_csv('/content/orders.csv')
df
```

```python
df.info()
```

```python
#Data_cleaning
```

```python
#renaming columns
df['Order Date']=pd.to_datetime(df['Order Date'], format='%Y-%m-%d',)
```

```python
df.info()
```

```python
#rename order id
df.rename(columns={
    'Order Id': 'order_id',
    'Order Date': 'order_date',
    'Ship Mode': 'ship_mode',
    'Segment': 'segment',
    'Country': 'country',
    'City': 'city',
    'State': 'state',
    'Region': 'region',
    'Category': 'category',
    'Postal Code': 'postal_code',
    'Sub Category': 'sub_category',
    'Product Id': 'product_id',
    'cost price': 'cost_price',
    'Quantity': 'quantity',
    'List Price': 'list_price',
    'Discount Percent': 'discount_percent'
}, inplace=True)
```

```python
df
```

```python
missing_values = df.isnull().sum()

# Display the results
print(missing_values)
```

```python
df.dropna(inplace=True)
```

```python
missing = df.isnull().sum()
print(missing)
```

```python
df
```

```python
df['sales_price']= (df['list_price'] - (df['list_price'] * (df['discount_percent'] / 100)))* df['quantity']
df['profit'] = df['sales_price'] - (df['cost_price'])* df['quantity']
```

```python
df['discount']= (df['list_price'])*(df['quantity']) - df['sales_price']
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['year'] = df['order_date'].dt.year
```

```python
df
```

```python
df.to_csv('retail_sales.csv',index=False)
```

```python
df1=df[['order_id','order_date','year','ship_mode','segment','country','city','state','region','category','sub_category','postal_code',]]
df2=df[['order_id','product_id','cost_price','list_price','discount_percent','sales_price','profit','discount','quantity']]
```

```python
df1.to_csv('product_details.csv',index=False)
df2.to_csv('sales.csv',index=False)
```

```python
# 11. Count the number of orders in each year.
query="""
SELECT pd.year, COUNT(s.order_id) AS total_orders
FROM sales s
JOIN product_details pd ON s.order_id = pd.order_id
GROUP BY pd.year
ORDER BY pd.year DESC;"""
cursor.execute(query)
rows = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
data10 = pd.DataFrame(rows, columns=columns)
px.pie(data10,names='year',values='total_orders')
```

```python
#12 Determine the most popular category by total sales quantity.
query="""
SELECT pd.category, SUM(s.quantity) AS total_sales_quantity
FROM sales s
JOIN product_details pd ON s.order_id = pd.order_id
GROUP BY pd.category
ORDER BY total_sales_quantity DESC
LIMIT 1;"""
cursor.execute(query)
rows = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
data11 = pd.DataFrame(rows, columns=columns)
data11
```

```python
#13. Identify states with profit greater than $10,000.
query="""SELECT pd.state, SUM(s.profit) AS total_profit
FROM sales s
JOIN product_details pd ON s.order_id = pd.order_id
GROUP BY pd.state
HAVING SUM(s.profit) > 10000
ORDER BY total_profit DESC;"""
cursor.execute(query)
rows = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
data12 = pd.DataFrame(rows, columns=columns)
px.pie(data12,names='state',values='total_profit')
```

```python
#14. Compute the average list price for each sub-category.
query="""SELECT pd.sub_category, AVG(s.list_price) AS avg_list_price
FROM sales s
JOIN product_details pd ON s.order_id = pd.order_id
GROUP BY pd.sub_category
ORDER BY avg_list_price DESC;"""
cursor.execute(query)
rows = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
data13 = pd.DataFrame(rows, columns=columns)
px.histogram(data13,x='sub_category',y='avg_list_price',color='sub_category')
```

```python
# 15 List the total sales by ship mode.
query="""SELECT pd.ship_mode, SUM(s.sales_price) AS total_sales
FROM sales s
JOIN product_details pd ON s.order_id = pd.order_id
GROUP BY pd.ship_mode
ORDER BY total_sales DESC;"""
cursor.execute(query)
rows = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
data14 = pd.DataFrame(rows, columns=columns)
px.bar(data14,x='ship_mode',y='total_sales',color='ship_mode')
```

```python
# 16. List 5 cities with the lowest average profit.
query="""SELECT pd.city, AVG(s.profit) AS avg_profit
FROM sales s
JOIN product_details pd ON s.order_id = pd.order_id
GROUP BY pd.city
ORDER BY avg_profit ASC
LIMIT 5;"""
cursor.execute(query)
rows = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]
data15 = pd.DataFrame(rows, columns=columns)
px.bar(data15,x='city',y='avg_profit',color='city')
```
