```python
#renaming columns
df['Order Date']=pd.to_datetime(df['Order Date'], format='%Y-%m-%d',)
```

```python
#rename order id
df.rename(columns={
    'Order Id': 'order_id',
    'Order Date': 'order_date',
    'Ship Mode': 'ship_mode',
    'Segment': 'segment',
    'Country': 'country',
    'City': 'city',
    'State': 'state',
    'Region': 'region',
    'Category': 'category',
    'Postal Code': 'postal_code',
    'Sub Category': 'sub_category',
    'Product Id': 'product_id',
    'cost price': 'cost_price',
    'Quantity': 'quantity',
    'List Price': 'list_price',
    'Discount Percent': 'discount_percent'
}, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['sales_price']= (df['list_price'] - (df['list_price'] * (df['discount_percent'] / 100)))* df['quantity']
df['profit'] = df['sales_price'] - (df['cost_price'])* df['quantity']
```

```python
df['discount']= (df['list_price'])*(df['quantity']) - df['sales_price']
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
```

```python
filtered_df = df[(df['quantity'] * df['discount_percent']) > 20]
print(filtered_df)
```
