```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python

df=pd.read_csv('/content/orders.csv')
df
```

```python
df.info()
```

```python
#Data_cleaning
```

```python
#renaming columns
df['Order Date']=pd.to_datetime(df['Order Date'], format='%Y-%m-%d',)
```

```python
df.info()
```

```python
#rename order id
df.rename(columns={
    'Order Id': 'order_id',
    'Order Date': 'order_date',
    'Ship Mode': 'ship_mode',
    'Segment': 'segment',
    'Country': 'country',
    'City': 'city',
    'State': 'state',
    'Region': 'region',
    'Category': 'category',
    'Postal Code': 'postal_code',
    'Sub Category': 'sub_category',
    'Product Id': 'product_id',
    'cost price': 'cost_price',
    'Quantity': 'quantity',
    'List Price': 'list_price',
    'Discount Percent': 'discount_percent'
}, inplace=True)
```

```python
df
```

```python
missing_values = df.isnull().sum()

# Display the results
print(missing_values)
```

```python
df.dropna(inplace=True)
```

```python
missing = df.isnull().sum()
print(missing)
```

```python
df
```

```python
df['sales_price']= (df['list_price'] - (df['list_price'] * (df['discount_percent'] / 100)))* df['quantity']
df['profit'] = df['sales_price'] - (df['cost_price'])* df['quantity']
```

```python
df['discount']= (df['list_price'])*(df['quantity']) - df['sales_price']
```

```python
df
```

```python
df.drop(columns=['Discount'],inplace=True)
```

```python
df.to_csv('retail_sales.csv',index=False)
```

```python
df1=df[['order_id','order_date','ship_mode','segment','country','city','state','region','category','sub_category','postal_code','quantity']]
df2=df[['order_id','product_id','cost_price','list_price','discount_percent','sales_price','profit','discount']]
```

```python
df1.to_csv('product_details.csv',index=False)
df2.to_csv('sales.csv',index=False)
```
